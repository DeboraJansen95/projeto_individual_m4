```python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Dados fornecidos
dias = ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo']
horas_trabalhadas = [6, 7, 8, 6, 7, 5, 4]
bugs_corrigidos = [3, 2, 1, 4, 3, 2, 1]
tarefas_concluidas = [5, 4, 6, 4, 5, 3, 2]

# Criando um DataFrame com os dados
df = pd.DataFrame({
    'Dia': dias,
    'Horas Trabalhadas': horas_trabalhadas,
    'Bugs Corrigidos': bugs_corrigidos,
    'Tarefas Concluídas': tarefas_concluidas
})

# Calculando as métricas solicitadas
total_horas = df['Horas Trabalhadas'].sum()
media_horas = df['Horas Trabalhadas'].mean()
total_bugs = df['Bugs Corrigidos'].sum()
media_bugs = df['Bugs Corrigidos'].mean()
total_tarefas = df['Tarefas Concluídas'].sum()
media_tarefas = df['Tarefas Concluídas'].mean()
produtividade = total_tarefas / total_horas

# Criando um DataFrame com as métricas calculadas
df_metricas = pd.DataFrame({
    'Métrica': ['Total de Horas', 'Média de Horas', 'Total de Bugs', 'Média de Bugs', 'Total de Tarefas', 'Média de Tarefas', 'Produtividade'],
    'Valor': [total_horas, media_horas, total_bugs, media_bugs, total_tarefas, media_tarefas, produtividade]
})

# Exibindo o relatório em formato de tabela
display(df_metricas)

# Criando gráficos para visualizar os dados
plt.figure(figsize=(16, 12))

# Gráfico de Linhas para Horas Trabalhadas
plt.subplot(2, 2, 1)
plt.plot(df['Dia'], df['Horas Trabalhadas'], marker='o')
plt.xlabel('Dia')
plt.ylabel('Horas Trabalhadas')
plt.title('Horas Trabalhadas por Dia')

# Gráfico de Barras Empilhadas para Tarefas Concluídas e Bugs Corrigidos
plt.subplot(2, 2, 2)
plt.bar(df['Dia'], df['Tarefas Concluídas'], label='Tarefas Concluídas')
plt.bar(df['Dia'], df['Bugs Corrigidos'], bottom=df['Tarefas Concluídas'], label='Bugs Corrigidos')
plt.xlabel('Dia')
plt.ylabel('Quantidade')
plt.title('Tarefas Concluídas e Bugs Corrigidos por Dia')
plt.legend()

# Gráfico de Dispersão para Horas Trabalhadas vs Tarefas Concluídas
plt.subplot(2, 2, 3)
plt.scatter(df['Horas Trabalhadas'], df['Tarefas Concluídas'])
plt.xlabel('Horas Trabalhadas')
plt.ylabel('Tarefas Concluídas')
plt.title('Horas Trabalhadas vs Tarefas Concluídas')

# Gráfico de Barras para as Métricas Calculadas
metricas = ['Total de Horas', 'Média de Horas', 'Total de Bugs', 'Média de Bugs', 'Total de Tarefas', 'Média de Tarefas', 'Produtividade']
valores = [total_horas, media_horas, total_bugs, media_bugs, total_tarefas, media_tarefas, produtividade]
y_pos = np.arange(len(metricas))
plt.subplot(2, 2, 4)
plt.barh(y_pos, valores, align='center')
plt.yticks(y_pos, metricas)
plt.xlabel('Valor')
plt.title('Visualização das Métricas Calculadas')

# Ajustando o layout e exibindo os gráficos
plt.tight_layout()
plt.show()

```
